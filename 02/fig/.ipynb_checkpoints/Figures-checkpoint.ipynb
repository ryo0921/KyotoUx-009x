```python
% matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot') # use "ggplot" style for graphs
```

```python
#plt.hist(R, bins=100,normed=True)    # R1001
fig, ax = plt.subplots(figsize=(7.5,7.5))
plt.xlim(-5, 5)
plt.ylim(-0.05,0.85)
x = np.arange(-5, 5, 0.01)           # -550.01x
###
std = 0.5                            # 
ave = 0.0                            # 
y = np.exp(-(x-ave)**2/2/std**2)/np.sqrt(2*np.pi*std**2) # xy
plt.plot(x, y, lw=5, color='k')      # x vs. y2
###
std = 1.0                            # 
ave = 0.0                            # 
y = np.exp(-(x-ave)**2/2/std**2)/np.sqrt(2*np.pi*std**2) # xy
plt.plot(x, y, lw=5, color='r')      # x vs. y2
###
std = 2.0                            # 
ave = 0.0                            # 
y = np.exp(-(x-ave)**2/2/std**2)/np.sqrt(2*np.pi*std**2) # xy
plt.plot(x, y, lw=5, color='b')      # x vs. y2
###
plt.xlabel(r'$x-\langle X \rangle$', fontsize=28)
plt.ylabel(r'$P(x)$', fontsize=28)
plt.legend([r'$\sigma=0.5$',r'$\sigma=1.0$',r'$\sigma=2.0$'], fontsize=24)
plt.tick_params(labelsize=20)
plt.show()                           # 1
```

```python
x = np.arange(M)                                      # 0M1x
y = np.zeros(M)                                       # y0
fig, ax = plt.subplots(figsize=(7.5,7.5))
plt.xlim(0, 100)
plt.ylim(-0.01,0.14)
#plt.ylim(0, 100)
def binomial(n,m,p):
    comb = math.factorial(m) / (math.factorial(n) * math.factorial(m-n))
    prob = comb * p ** n * (1 - p) ** (m - n)
    return prob
p = 0.5                                               # PQP
M = 50                                               # 
for i in range(M):
    y[i]=binomial(i,M,p)                              # y
plt.plot(x, y, lw=4, color='k')                       # x vs. y2
p = 0.5                                               # PQP
M = 100                                               # 
for i in range(M):
    y[i]=binomial(i,M,p)                              # y
plt.plot(x, y, lw=4, color='r')                       # x vs. y2
p = 0.7                                               # PQP
M = 100                                               # 
for i in range(M):
    y[i]=binomial(i,M,p)                              # y
plt.plot(x, y, lw=4, color='b')                       # x vs. y2

#ax = plt.axes()
#plt.title(r'Binomial Distribution: p=%f, M=%i' % (p,M))
plt.xlabel(r'$n$', fontsize=28)
plt.ylabel(r'$P(n)$', fontsize=28)
ax.legend([r'$a=0.5,M=50$',r'$a=0.5,M=100$',r'$a=0.7,M=100$'], fontsize=24)
plt.tick_params(labelsize=20)
plt.show()                                             # 1
```

```python
fig, ax = plt.subplots(figsize=(7.5,7.5))
plt.xlim(0, 20)
plt.ylim(-0.01, 0.4)
M = 100                                               # 
x = np.arange(M)                                      # 0M1x
y = np.zeros(M)                                       # y0

def Poisson(a,n):
    prob = a**n*np.exp(-a)/math.factorial(n)
    return prob
###
a = 1                                               # PQP
for i in range(M):
    y[i]=Poisson(a,i)                              # y
plt.plot(x, y, lw=4, color='k')                       # x vs. y2
###
a = 5                                               # PQP
for i in range(M):
    y[i]=Poisson(a,i)                              # y
plt.plot(x, y, lw=4, color='r')                       # x vs. y2
###
a = 10                                               # PQP
for i in range(M):
    y[i]=Poisson(a,i)                              # y
plt.plot(x, y, lw=4, color='b')                       # x vs. y2
###
#ax = plt.axes()
#plt.title(r'Binomial Distribution: p=%f, M=%i' % (p,M))
plt.xlabel(r'$n$', fontsize=28)
plt.ylabel(r'$P(n)$', fontsize=28)
ax.legend([r'$a=1$',r'$a=5$',r'$a=10$'], fontsize=24)
plt.tick_params(labelsize=20)
plt.show()                                             # 1
```

```python
fig, ax = plt.subplots(figsize=(7.5,5))
def binomial(n,m,p):
    comb = math.factorial(m) / (math.factorial(n) * math.factorial(m-n))
    prob = comb * p ** n * (1 - p) ** (m - n)
    return prob
p = 0.5                                               # PQP
M = 20                                               # 
N = 100000                                            # 
np.random.seed(0)                                     # 
R=np.random.binomial(n=M, p=p, size=N)                # MPRN
plt.xlabel(r'$i$', fontsize=28)
plt.ylabel(r'$x_i$', fontsize=28)
plt.xlim(0,1000)
#plt.ylim(-5,5)
plt.tick_params(labelsize=20)
plt.plot(R)
plt.show()
```
